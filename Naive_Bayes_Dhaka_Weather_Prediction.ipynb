```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
```

```python
df = pd.read_csv('/content/weather_history_bangladesh.csv')

df

```

Remove Duplicate data

```python


# Remove duplicate rows
df = df.drop_duplicates()

# Display the dataset after removing duplicates
df
```

```python
Numerics = LabelEncoder()
```

```python
print(df.columns)
```

```python
inputs= df.drop('condition', axis = 'columns')
target = df['condition']
target
```

```python
inputs['date_n']=Numerics.fit_transform(inputs['date'])
inputs['time_n']=Numerics.fit_transform(inputs['time'])
inputs['temperature_fahrenheit_n']=Numerics.fit_transform(inputs['temperature_fahrenheit'])
inputs['dew_point_fahrenheit_n']=Numerics.fit_transform(inputs['dew_point_fahrenheit'])
inputs['humidity_percentage_n']=Numerics.fit_transform(inputs['humidity_percentage'])
inputs['wind_n']=Numerics.fit_transform(inputs['wind'])
inputs['wind_speed_mph_n']=Numerics.fit_transform(inputs['wind_speed_mph'])
inputs['wind_gust_mph_n']=Numerics.fit_transform(inputs['wind_gust_mph'])
inputs['pressure_in_n']=Numerics.fit_transform(inputs['pressure_in'])
inputs['precip._in_n']=Numerics.fit_transform(inputs['precip._in'])
inputs
```

```python
inputs_n = inputs.drop(['date', 'time', 'temperature_fahrenheit', 'dew_point_fahrenheit',
       'humidity_percentage', 'wind', 'wind_speed_mph', 'wind_gust_mph',
       'pressure_in', 'precip._in'], axis = 'columns')
inputs_n
```

```python
Classifier = GaussianNB()
Classifier.fit(inputs_n, target)

```

```python
# Calculate accuracy
accuracy = Classifier.score(inputs_n, target)

# Print accuracy as a percentage with two decimal places
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
# Preprocessing
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
inputs_scaled = scaler.fit_transform(inputs_n)

# Train-Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs_scaled, target, test_size=0.2, random_state=42)

# Train Classifier
Classifier = GaussianNB(var_smoothing=1e-8)
Classifier.fit(X_train, y_train)

# Evaluate Model
accuracy = Classifier.score(X_test, y_test)
print(f"Test Accuracy: {accuracy * 100:.2f}%")
```
