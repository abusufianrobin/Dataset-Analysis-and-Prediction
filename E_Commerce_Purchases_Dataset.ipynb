14. How many people have a credit card that expires in 2020?

```python
# Q No 14: How many people have a credit card that expires in 2020?

def function():
  count=0
  for date in data['CC Exp Date']:
   if date.split('/')[1]=='20':
    count = count+1
  print("Number of People have Expired Credit Card:",count)

function()

# Another Method

# Using len() method

#len(data[data['CC Exp Date'].apply(lambda x:x[3: ]=='20')])
```

15. What are the top 5 most popular email providers (e.g. gmail.com, yahoo.com, etc...)


```python
# Q No 15: What are the top 5 most popular email providers (e.g. gmail.com, yahoo.com, etc...)

list1 = []
for email in data['Email']:
  list1.append(email.split('@')[1])

data['Temp'] = list1
data.head(1)

data['Temp'].value_counts().head()

# Another Method: Using lambda() method

#data['Email'].apply(lambda x:x.split('@')[1]).value_counts().head()
```
