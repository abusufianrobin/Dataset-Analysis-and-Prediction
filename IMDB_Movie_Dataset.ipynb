20. Classify Movies Based on Ratings [Excellent, Good, and Average]

```python
# Q No 20: Classify Movies Based on Ratings [Excellent, Good, and Average]

data.columns

def rating(rating):
  if rating >= 7.0:
    return "Excellent"

  elif rating >= 6.0:
    return "Good"

  else:
    return "Average"

data['Rating_Category'] = data['Rating'].apply(rating)

data.head(5)
```

21. Count Number of Action Movies

22. Find Unique Values From Genre

```python
# Q No 22: Find Unique Values From Genre

data['Genre']

# 1st step:

list1 = []

for value in data['Genre']:
  list1.append(value.split(','))

list1

```

```python
# 2nd Step:

one_d = []
for item in list1:
  for item1 in item:
    one_d.append(item1)

one_d
```

```python
#3rd Step:

uni_list = []

for item in one_d:
  if item not in uni_list:
    uni_list.append(item)

uni_list

#len(uni_list)
```

23. How Many Films of Each Genre Were Made?

```python
# Q No 23: How Many Films of Each Genre Were Made?

one_d = []
for item in list1:
  for item1 in item:
    one_d.append(item1)

one_d

from collections import Counter

Counter(one_d)
```

For Visualize the data

```python
# Diagram for visualization

genre_counts = Counter(one_d)

# Convert the counter object to a DataFrame for easier plotting
genre_df = pd.DataFrame.from_dict(genre_counts, orient='index').reset_index()
genre_df.columns = ['Genre', 'Count']

# Sort the DataFrame by count (optional, for better visualization)
genre_df = genre_df.sort_values(by='Count', ascending=False)

# Create a barplot
plt.figure(figsize=(10, 8))
sns.barplot(x='Genre', y='Count', data=genre_df, palette='viridis')

# Add title and labels
plt.title("Number of Films by Genre", fontsize=16)
plt.xlabel("Genre", fontsize=12)
plt.ylabel("Number of Movies", fontsize=12)

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```
