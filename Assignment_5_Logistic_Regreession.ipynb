```python
import numpy as np
import matplotlib.pyplot as plt


def sigmoid(z):
    return 1 / (1 + np.exp(-z))


def compute_cost(X, y, weights):
    n_samples = len(y)
    predictions = sigmoid(np.dot(X, weights))
    cost = -(1/n_samples) * np.sum(y * np.log(predictions) + (1 - y) * np.log(1 - predictions))
    return cost
```

```python
data = np.genfromtxt('emails.csv', delimiter=',', skip_header=1)


# Extract X and y (features and labels)
X = data[:, 1:-1]
y = data[:, -1]

# Normalize the dataset
X = (X - np.mean(X, axis=0)) / np.std(X, axis=0)

# Add bias term (column of ones)
X = np.c_[X, np.ones(X.shape[0])]

train_size = int(0.8 * len(X))
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

Learning Rate (lr) = 0.1;


```python
class LogisticRegression:
    def __init__(self, lr=0.1, n_iters=500):
        self.lr = lr
        self.n_iters = n_iters
        self.weights = None
        self.history = []

    def fit(self, X, y):
        n_samples, n_features = X.shape

        self.weights = np.zeros(n_features)

        for i in range(self.n_iters):

            z = np.dot(X, self.weights)

            predictions = sigmoid(z)


            dw = (1 / n_samples) * np.dot(X.T, (predictions - y))
            self.weights -= self.lr * dw


            cost = compute_cost(X, y, self.weights)
            self.history.append(cost)

    def predict(self, X):
        z = np.dot(X, self.weights)
        predictions = sigmoid(z)
        return [1 if i >= 0.5 else 0 for i in predictions]


model = LogisticRegression(lr=0.1, n_iters=500)
model.fit(X_train, y_train)


plt.plot(range(1, model.n_iters + 1), model.history, label='Learning Rate: 0.1')
plt.title("Total Cost vs Iteration")
plt.xlabel("Iteration")
plt.ylabel("Cost")
plt.legend()
plt.show()


y_pred = model.predict(X_test)


accuracy = np.sum(y_pred == y_test) / len(y_test)
print(f"Accuracy on the test set: {accuracy * 100:.2f}%")
```
